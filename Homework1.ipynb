```python
import matplotlib.pyplot as plt
from math import sin, pi
import numpy as np


def f(x, n):
    sums = 0
    for i in range (1,n,2):
        sums += sin(i*x)/i
#        print(sin(i*x)/i)
#    print(sums)
    return pi/2 + 2*(sums)
x = np.linspace(-pi,pi,num=50)

plt.plot(x,[f(i,31) for i in x])
plt.show()
```
